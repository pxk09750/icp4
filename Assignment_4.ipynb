```python
import pandas
from matplotlib import pyplot as plt

data_frame = pandas.read_csv('data.csv')

description = data_frame.describe()
print("                DATAFRAME DESCRIPTION  ")
print("")
print(description)
print("")
null_columns = data_frame.columns[data_frame.isnull().any()]
for c in null_columns:
   mean_value = data_frame[c].mean()
   data_frame[c].fillna(mean_value, inplace=True)

selected_columns = ['Calories', 'Duration']
aggregated_data = data_frame[selected_columns].agg(['min', 'max', 'count', 'mean'])
print("         AGGREGATED DATA  ")
print("")
print(aggregated_data)
print("")

filtered_df1 = data_frame[(data_frame['Calories'] >= 500) & (data_frame['Calories'] <= 1000)]

filtered_df2 = data_frame[(data_frame['Calories'] > 500) & (data_frame['Pulse'] < 100)]

df_modified = data_frame.drop(columns=['Maxpulse'])
data_frame.drop(columns=['Maxpulse'], inplace=True)
data_frame['Calories'] = data_frame['Calories'].astype(int)

plt.scatter(data_frame['Duration'], data_frame['Calories'])
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.title('DURATION VS CALORIES GRAPH')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot as plt

data = pd.read_csv("Salary_Data.csv")

x = data[['YearsExperience']]
y = data[['Salary']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)

regressor = LinearRegression()
regressor.fit(x_train, y_train)

y_pred = regressor.predict(x_test)

mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error:", mse)

plt.scatter(x_train, y_train, label="Training Data")
plt.scatter(x_test, y_test, label="Test Data")
plt.plot(x_test, y_pred, label="Linear Regression")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.title("Linear Regression: Salary vs Years of Experience")
plt.show()
```
